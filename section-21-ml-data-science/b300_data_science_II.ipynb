We learned about importing the data and using the Pandas library to manipulate our data frames. How do we __clean__ the data?

Boss wants to know which players are a good deal: most skilled and lowest in salary. So he wants to know current market value and the wage: that is, which player is undervalued right now. We'll grab the following

- Value
- Wage
- Name

```python
import pandas as pd
data_frame = pd.read_csv('./soccer-data/data.csv')
data_frame[data_frame["Age"]>40].head()
```

```python
df1 = pd.DataFrame(data_frame, columns=['Name', 'Wage', 'Value'])
df1 
```

Just like that, we have the qualities of the player we're looking for. But maybe the data doesn't make the sense that we want it to. Let's sort it.

We want to compare each player's wage to the value and have a different column called `difference`. We want to see which player has the biggest gap between their `'Wage'` and their `'value'`.

Ideally, we can just create a `difference` column by saying:

```python
df1 = pd.DataFrame(data_frame, columns=['Name', 'Wage', 'Value'])
df1['difference'] = df1['Value'] - df1['Wage']
df1
```

The error, `unsupported operand type(s) for -: 'str' and 'str'`, tells us we're trying to extract a string and a string. The wage of a player is in Euros and we need to convert this into a float.

```python
df1 = pd.DataFrame(data_frame, columns=['Name', 'Wage', 'Value'])
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

wage = df1['Wage'].replace('[\€,]', '', regex=True).apply(value_to_float)
value = df1['Value'].replace('[\€,]', '', regex=True).apply(value_to_float)

df1['Wage'] = wage
df1['Value'] = value

df1['difference'] = df1['Value'] - df1['Wage']
df1
```

```python
df1.sort_values(by='difference', ascending=False).head(5)
```

Here, we get the players with the biggest difference down to the lowest players.
